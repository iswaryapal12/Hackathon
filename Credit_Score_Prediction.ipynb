```python
credit_train=credit[credit["Data"]=="Train"]
credit_test=credit[credit["Data"]=="Test"]
```

```python
del credit_train["Data"]
del credit_test["Data"]
```

```python
x_train=credit_train.drop("y",axis=1)
y_train=credit_train["y"]

x_test=credit_test.drop("y",axis=1)
y_test=credit_test["y"]
```

```python
from sklearn.linear_model import LinearRegression
linreg = LinearRegression()
linreg.fit(x_train, y_train)
train_lin = linreg.predict(x_train)
test_lin = linreg.predict(x_test)
```

```python
#Find all the errors
from math import sqrt
def general_errors(train, pred):
    mae=mean_absolute_error(train, pred)
    mse=mean_squared_error(train, pred)
    rmse = sqrt(mean_squared_error(train, pred))
    print("MAE: ",mae,"||","MSE: ",mse,"||","RMSE: ",rmse)
    return
```

```python
#MAPE errors
def mape(y_train, train_Pred): 
    y_true, y_pred = np.array(y_train), np.array(train_Pred)
    print("MAPE: ",np.mean(np.abs((y_true - y_pred) / y_true)) * 100)
    return
```

```python
#train data
general_errors(y_train, train_lin)
mape(y_train, train_lin)
```

```python
#test data
general_errors(y_test, test_lin)
mape(y_test, test_lin)
```
